## Gráfico da idade

Podemos ver que o codigo abaixo produz o grafico de barras para cada idade separada
Os passos são:
* Abrimos o arquivo
* Somamos o numero de pessoas por cada idade (separando-as por um dicionário)
* Criamos uma lista com os elementos que futuramente serão o eixo x do meu gráfico
* Printamos os elementos do arquivo com suas respectivas quantidades
* Criamos a lista com o eixo Y (quantidades de cada grupo)
* Finalmente fazemos o gráfico de barras
* Conferimos o tempo para ter uma noção da eficiência do algorítmo

```python
import csv
import time
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
```

```python
%matplotlib notebook

inicio = time.time()

with open("perfil_eleitorado_ATUAL.csv") as arq:
    dados = csv.reader(arq,delimiter = ';')
    dici = {}
    for linhas in dados:
        if linhas[6] not in dici:
            dici[linhas[6]] = 0
        dici[linhas[6]] += int(linhas[8])

lista = ['16','17','18 A 20','21 A 24','25 A 34','35 A 44','45 A 59','60 A 69','70 A 79','SUPERIOR A 79','INVÁLIDA']
lista2 = []
print("VALORES ABSOLUTOS: \n")
for i in dici:
    print(i,":  ",dici[i])
    lista2.append(dici[i])

plt.bar(lista,lista2,color = 'c')


fim = time.time()
print("\n"*10,"Tempo de duração do processo: ",fim-inicio)
```
