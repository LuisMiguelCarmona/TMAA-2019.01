```python
## Luis Miguel Moura Carmona
## DRE - 119066153
## Matemática Aplicada

try:    
    tamanho = int(input("Qual é o tamanho dos blocos? "))
    mini = int(input("Qual é o menor numero da tabela ASCII que você quer? ")) -1
    maxi = int(input("Qual é o maior numero da tabela ASCII que você quer? "))
except:
    tamanho = 25
    mini = 31
    maxi = 127
Bina = []
while mini < maxi:
    K = 0
    print("\n+-----------+-----+-----+-----+----+")
    print("|  Binário  | Oct | Hex | Dec | Chr|")
    print("+-----------+-----+-----+-----+----+")
    while K < tamanho:
        mini += 1
        if mini > maxi:
            break
        else:
            binario = ("{0:0b}".format(mini)).zfill(8)
            hexa = ("{0:3X}".format(mini))
            octa = ("{0:2o}".format(mini)).zfill(3)
            deci = ("{0:2}".format(mini)).zfill(3)
            Bina = list((str(binario)))
            print("|","".join(Bina[0:4]),"".join(Bina[4:8]),"|",octa,"|",hexa,"|",deci,"| ",chr(mini),"|")
            K += 1
    print("+-----------+-----+-----+-----+----+")
```
