```python
## Luis Miguel Moura Carmona
## DRE - 119066153
## Matemática Aplicada


## Tabela em braile está completa (não apenas de 6 a 8 pontos)

try:    
    tamanho = int(input("Qual é o tamanho dos blocos? "))
    mini = int(input("Qual é a menor letra em braile (começando no 0)? ")) + 10240
    maxi = int(input("Qual é a maior letra em braile (começando no 0, fim em 254)? ")) + 10241
except:
    tamanho = 25
    mini = 10241
    maxi = 10495
Bina = []
while mini < maxi:
    K = 0
    print("\n+-----------------+-------+------+-------+-----+")
    print("|    Binário      |  Oct  |  Hex |  Dec  | Chr |")
    print("+-----------------+-------+------+-------+-----+")
    while K < tamanho:
        mini += 1
        if mini > maxi:
            break
        else:
            binario = ("{0:0b}".format(mini)).zfill(15)
            hexa = ("{0:3X}".format(mini))
            octa = ("{0:2o}".format(mini)).zfill(3)
            deci = ("{0:2}".format(mini)).zfill(3)
            Bina = list((str(binario)))
            print("|","".join(Bina[0:7]),"".join(Bina[7:-1]),"|",octa,"|",hexa,"|",deci,"| ",chr(mini)," |")
            K += 1
    print("+-----------------+-------+------+-------+-----+")
```
