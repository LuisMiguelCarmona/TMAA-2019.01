```python
## Luis Miguel Moura Carmona
## DRE - 119066153
## Matemática Aplicada

try:    
    tamanho = int(input("Qual é o tamanho dos blocos? "))
    mini = int(input("Qual é o letra do grego (começa no 0, vai até o 57)? ")) +912
    maxi = int(input("Qual é a maior letra do grego (começa no 0, vai até o 57)? ")) +913
except:
    tamanho = 25
    mini = 912
    maxi = 969
Bina = []
while mini < maxi:
    K = 0
    print("\n+-------------+------+-----+-----+----+")
    print("|   Binário   | Oct  | Hex | Dec | Chr|")
    print("+-------------+------+-----+-----+----+")
    while K < tamanho:
        mini += 1
        if mini > maxi:
            break
        else:
            binario = ("{0:0b}".format(mini)).zfill(11)
            hexa = ("{0:3X}".format(mini))
            octa = ("{0:2o}".format(mini)).zfill(3)
            deci = ("{0:2}".format(mini)).zfill(3)
            Bina = list((str(binario)))
            print("|","".join(Bina[:5]),"".join(Bina[5:-1]),"|",octa,"|",hexa,"|",deci,"| ",chr(mini),"|")
            K += 1
    print("+------------+------+-----+-----+----+")
```
