```python
## Luis Miguel Carmona
## DRE - 119066153
## Matemática Aplicada

import os
def PastadosArq(arq='.'):
    arquivos=os.listdir(arq)
    textos=[]
    for txt in arquivos:
        if txt[-4:]=='.txt':
            textos.append(txt)
    palavras=[]
    freq1=[]
    final=[]
    print(textos)
    for i in textos:
        arquivo=open(i,'r')
        q=arquivo.read().lower() ## Transformamos tudo em minusculo para evitar palavras que iniciam paragrafos
        separado=q.split(" ")
        for p in separado:
            p.replace(',',' ')
            p.replace('.',' ')
            p.replace(':',' ')
        palavras+=separado
    for i in palavras:
        separado=i.split('\n')
        palavras.remove(i)
        palavras+=separado
    freq1 =list(dict.fromkeys(palavras))
    for plv in freq1:
        contag=str(palavras.count(plv))
        final.append(plv+' - '+contag)
    final.sort() ## Pela ordem Lexical, caso queira pela ordem numerica só criar um lambda
    return final

PastadosArq()
```
